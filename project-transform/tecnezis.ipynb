Информация о количестве совпадающих названий по городу и названию платформы 2

```python
def data_user_num(col): 
    y = [0]
    for x in col:
        if x > 0:
            y.append(x)
    z = y[-1]
    return z
```

```python
def data_user_str(col):
    y = ['']
    for x in col:
        y.append(x)
    z = y[-1]
    return z
```

Созданы функции для обработки записей. *data_user_num* для обработки числовых записей, *data_user_str* для обработки данных типа object. 
1. Учитывается, что записи могут повторятся, и самые актуальные данные идут в конце. 
2. Если какое-то поле по конкретной гостинице пустое в последней записи, но встречалось ранее, то оно взято из более ранних записей

```python
data1['rating'] = data1.groupby(['title','address'])['rating'].transform(lambda x: data_user_num(x))
data1['rating_5'] = data1.groupby(['title','address'])['rating_5'].transform(lambda x: data_user_num(x))
data1['review_count'] = data1.groupby(['title','address'])['review_count'].transform(lambda x: data_user_num(x))
data1['rooms_count'] = data1.groupby(['title','address'])['review_count'].transform(lambda x: data_user_num(x))
data1['star_rating'] = data1.groupby(['title','address'])['star_rating'].transform(lambda x: data_user_num(x))
data1['lat'] = data1.groupby(['title','address'])['lat'].transform(lambda x: data_user_num(x))
data1['lon'] = data1.groupby(['title','address'])['lon'].transform(lambda x: data_user_num(x))
data1['description'] = data1.groupby(['title','address'])['description'].transform(lambda x: data_user_str(x))
data1['phone'] = data1.groupby(['title','address'])['phone'].transform(lambda x: data_user_str(x))
data1['email'] = data1.groupby(['title','address'])['email'].transform(lambda x: data_user_str(x))
data1['website'] = data1.groupby(['title','address'])['website'].transform(lambda x: data_user_str(x))
data1['contact_social'] = data1.groupby(['title','address'])['contact_social'].transform(lambda x: data_user_str(x))
data1['city'] = data1.groupby(['title','address'])['city'].transform(lambda x: data_user_str(x))
data1['hotel_type_original'] = data1.groupby(['title','address'])['hotel_type_original'].transform(lambda x: data_user_str(x))
```

Выявлен город с наибольшим количеством гостиниц в платформе 1

```python
# создаём карту c стартовыми координатами Калиниграда
start_lat =  54.716667
start_lon = 20.516667
m = Map(location=[start_lat, start_lon], zoom_start=13, control_scale = True ,tiles="Cartodb Positron")
# создаём пустой кластер, добавляем его на карту
marker_cluster = MarkerCluster().add_to(m)

# пишем функцию, которая принимает строку датафрейма,
# создаёт маркер в текущей точке и добавляет его в кластер marker_cluster
def create_clusters(row):
    Marker(
        [row['lat'], row['lon']]
        
    ).add_to(marker_cluster)

# применяем функцию create_clusters() к каждой строке датафрейма
data1.apply(create_clusters, axis=1)

# выводим карту
m
```

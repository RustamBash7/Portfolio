Информация о количестве дубликатов по uid платформы 2

```python
def data_user_num(col): 
    y = [np.nan]
    for x in col:
        if pd.notna(x) :
            y.append(x)
    z = y[-1]
    return z
```

```python
def data_user_str(col):
    y = [np.nan]
    for x in col:
        if pd.notna(x) :
            y.append(x)
    z = y[-1]
    return z
```

Созданы функции для обработки записей. *data_user_num* для обработки числовых записей, *data_user_str* для обработки данных типа object. 
1. Учитывается, что записи могут повторятся, и самые актуальные данные идут в конце. 
2. Если какое-то поле по конкретной гостинице пустое в последней записи, но встречалось ранее, то оно взято из более ранних записей

```python
data2['rating'] = data2.sort_values(by=['uid','create_time']).groupby('uid')['rating'].transform(lambda x: data_user_num(x))
data2['rating_5'] = data2.sort_values(by=['uid','create_time']).groupby('uid')['rating_5'].transform(lambda x: data_user_num(x))
data2['review_count'] = data2.sort_values(by=['uid','create_time']).groupby('uid')['review_count'].transform(lambda x: data_user_num(x))
data2['rooms_count'] = data2.sort_values(by=['uid','create_time']).groupby('uid')['rooms_count'].transform(lambda x: data_user_num(x))
data2['star_rating'] = data2.sort_values(by=['uid','create_time']).groupby('uid')['star_rating'].transform(lambda x: data_user_num(x))
data2['lat'] = data2.sort_values(by=['uid','create_time']).groupby('uid')['lat'].transform(lambda x: data_user_num(x))
data2['lon'] = data2.sort_values(by=['uid','create_time']).groupby('uid')['lon'].transform(lambda x: data_user_num(x))
data2['description'] = data2.sort_values(by=['uid','create_time']).groupby('uid')['description'].transform(lambda x: data_user_str(x))
data2['phone'] = data2.sort_values(by=['uid','create_time']).groupby('uid')['phone'].transform(lambda x: data_user_str(x))
data2['email'] = data2.sort_values(by=['uid','create_time']).groupby('uid')['email'].transform(lambda x: data_user_str(x))
data2['website'] = data2.sort_values(by=['uid','create_time']).groupby('uid')['website'].transform(lambda x: data_user_str(x))
data2['contact_social'] = data2.sort_values(by=['uid','create_time']).groupby('uid')['contact_social'].transform(lambda x: data_user_str(x))
data2['city'] = data2.sort_values(by=['uid','create_time']).groupby('uid')['city'].transform(lambda x: data_user_str(x))
data2['hotel_type_original'] = data2.sort_values(by=['uid','create_time']).groupby('uid')['hotel_type_original'].transform(lambda x: data_user_str(x))
```

Обработка группированных по названию и адресу актуальных полей датафреймов с помощью функций. Актуальные - это поля необходимые для создания наиболее полной записи об гостиницах для дальнейшего исследования, поля с пропусками или нулевыми значениями (для рейтингов, зведности и количества номеров)

Выявлен город с наибольшим количеством гостиниц в платформе 1

```python
# создаём карту c стартовыми координатами Калиниграда
start_lat =  54.716667
start_lon = 20.516667
m = Map(location=[start_lat, start_lon], zoom_start=13, control_scale = True ,tiles="Cartodb Positron")
# создаём пустой кластер, добавляем его на карту
marker_cluster = MarkerCluster().add_to(m)

# пишем функцию, которая принимает строку датафрейма,
# создаёт маркер в текущей точке и добавляет его в кластер marker_cluster
def create_clusters(row):
    Marker(
        [row['lat'], row['lon']]
        
    ).add_to(marker_cluster)

# применяем функцию create_clusters() к каждой строке датафрейма
data1.apply(create_clusters, axis=1)

# выводим карту
m
```
